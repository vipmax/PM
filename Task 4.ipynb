```python
from datetime import datetime
import pandas as pd

converters = { 0: lambda x: datetime.strptime(x, '%Y-%m-%d'),  'Close': lambda x: float(x)}
FTSE_data = pd.read_csv("FTSE(input).csv", index_col = 0,  converters=converters , delimiter=',')
# print(RFSBL_data)

def calculate_relative_difference(pair):
    return (pair[0]-pair[1])/pair[1]

FTSE_data = pd.rolling_apply(FTSE_data['Close'], window=2, func=lambda pair: (pair[0]-pair[1])/pair[1])

FTSE = FTSE_data.resample('1d')
FTSE.values

# cov(FTSE.values)
```

```python
import numpy

def cov(t,vals1,vals2):
    mean = 0
    N = len(vals1)
    
    def isNotZero(i,t):
        if numpy.isnan(vals1[i]) is not False and numpy.isnan(vals2[i+t]) is not False: return True
        else: return False

    def pairsCoutn(t): 
        return sum([1 if isNotZero(i,t) else 0 for i in range(0, N - t - 1)])
    

    return 1/pairsCoutn(t) * sum([(vals1[i] - mean)*(vals2[i+t] - mean) if isNotZero(i,t) else 0 for i in range(0, N - t - 1)])
```

```python

vals1 = FTSE.values
vals2 = RFSBL.values

for i in range(len(vals1)-1):
    if numpy.isnan(vals1[i]): 
        vals1[i] = 0
    if numpy.isnan(vals2[i]): 
        vals2[i] = 0
        
        
print(len(vals1),len(vals2))

y = [cov(t, vals1, vals2) for t in range(0, len(vals1)-1)]
x = [i for i in range(0, len(y) )]


plt.figure(figsize=(20, 10))
plt.title('covariance FTSE and RFSBL')
plt.xlabel('t')
plt.ylabel('covariance FTSE and RFSBL')
plt.plot(x, y, label="covariance")
plt.plot(x, [0 for i in range(0, len(y))], label="0")

plt.show()
    
    

plt.figure(figsize=(20, 10))
plt.title('covariance FTSE and RFSBL')
plt.xlabel('t')
plt.ylabel('covariance FTSE and RFSBL')
plt.plot(x, y, label="covariance")
plt.plot(x, [0 for i in range(0, len(y) )], label="")
plt.xlim(0,10)

plt.show()
```
